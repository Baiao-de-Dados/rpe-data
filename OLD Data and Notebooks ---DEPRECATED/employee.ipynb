**Caminho da pasta e output da pasta**

**Mostrando Arquivos encontrados**

```python
dfs = []
files_in_folder = os.listdir(folder_path) # Define the variable here
print(f"Número total de arquivos encontrados: {len(files_in_folder)}") # Now you can use the variable
print(f"Verificando o conteúdo da pasta: {folder_path}")
print(f"Arquivos encontrados: {files_in_folder}")
```

**Adicionando Arquivos a um Dataframe**

**Tambem verificando pois varios arquivos tavam com .xlsx ou .XLSX**

```python
# Itera sobre todos os arquivos na pasta especificada
for file in files_in_folder:
    print(f"Processando arquivo: {file}") # Adiciona um print para ver qual arquivo está sendo processado
    # Verifica se o arquivo termina com .xlsx, ignorando a caixa (uppercase/lowercase)
    if file.lower().endswith('.xlsx'):
        file_path = os.path.join(folder_path, file)
        try:
            df = pd.read_excel(file_path)
            dfs.append(df)
            print(f'Arquivo {file} lido com sucesso e adicionado à lista de dataframes.') # Adiciona print para confirmar a leitura e adição
        except Exception as e:
            print(f'Erro ao ler o arquivo {file}: {e}')
    else:
        print(f"Arquivo {file} não é um arquivo .xlsx (verificação case-insensitive), ignorando.") # Adiciona print para arquivos ignorados

print(f"Total de dataframes lidos: {len(dfs)}") # Adiciona print para o número total de dataframes lidos
```

**Concatena todos os arquivos em um dataframe só em um so xlsx**

```python
# Combina todos os dataframes em um único dataframe
if dfs:
    combined_rpa = pd.concat(dfs, ignore_index=True)

    # Certifica-se de que o diretório de saída existe
    output_dir = os.path.dirname(output_file)
    os.makedirs(output_dir, exist_ok=True)

    # Salva o dataframe combinado em um novo arquivo Excel
    combined_rpa.to_excel(output_file, index=False)
    print(f'Todos os arquivos foram combinados e salvos em {output_file}')
else:
    print("Nenhum arquivo .xlsx encontrado na pasta especificada.")
```

**Tinha alguns sobrenomes que estavam incompletos adicionei alguns genericos**

```python
import numpy as np
import random
# Function to replace 'Da' with a random real surname
def replace_da_surname(name):
    if "Da" in name:
        # List of example real surnames (can be expanded)
        real_surnames = ["Silva", "Santos", "Oliveira", "Souza", "Lima", "Pereira", "Ferreira", "Almeida", "Costa", "Rodrigues"]
        # Replace "Da" with a random surname
        return name.replace("Da", random.choice(real_surnames))
    return name
# Apply the function to the 'Nome' column
combined_rpa['Nome'] = combined_rpa['Nome'].apply(replace_da_surname)
```

**Adicionei ao EMAIL**

```python
# Split the 'Nome' column by space and take the last element as the surname
combined_rpa['Sobrenome'] = combined_rpa['Nome'].str.split().str[-1].str.lower()
# Replace ".da" in the 'Email' column with the respective surname from the new 'Sobrenome' column
combined_rpa['Email'] = combined_rpa.apply(lambda row: row['Email'].replace('.da', '.' + row['Sobrenome']), axis=1)
# Drop the temporary 'Sobrenome' column
combined_rpa = combined_rpa.drop(columns=['Sobrenome'])
```
