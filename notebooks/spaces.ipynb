# Task
Read the Excel files "/content/ciclos/ciclo1.xlsx", "/content/ciclos/ciclo2.xlsx", and "/content/ciclos/ciclo3.xlsx". For the columns 'name', 'a360_name', and 'referencia_name' in each file, replace the '.' character with a space.

## List files

### Subtask:
List the three Excel files within "/content/ciclos".


**Reasoning**:
List files within the directory "/content/ciclos" and filter for Excel files.



```python
import os

directory_path = "/content/ciclos"
all_files = os.listdir(directory_path)
excel_files = [os.path.join(directory_path, f) for f in all_files if f.endswith('.xlsx')]

if excel_files:
    print("Excel files found in /content/ciclos:")
    for file_path in excel_files:
        print(file_path)
else:
    print("No Excel files found in /content/ciclos.")
```

## Iterate through files

### Subtask:
Loop through each Excel file identified in the previous step.


**Reasoning**:
Iterate through the list of Excel files and print the file path for each file.



```python
for file_path in excel_files:
    print(f"Processing file: {file_path}")
```

## Read data

### Subtask:
Read the data from the Excel file into a pandas DataFrame.


**Reasoning**:
Read the Excel file into a pandas DataFrame and display the head.



```python
import pandas as pd

for file_path in excel_files:
    print(f"Processing file: {file_path}")
    df = pd.read_excel(file_path)
    display(df.head())
```

## Extract columns

### Subtask:
Extract the 'name', 'a360_name', and 'referencia_name' columns from the DataFrame.


**Reasoning**:
Extract the specified columns from each DataFrame and display the head.



```python
columns_to_extract = ['name', 'a360_name', 'referencia_name']

for file_path in excel_files:
    print(f"Processing file: {file_path}")
    df = pd.read_excel(file_path)
    extracted_columns_df = df[columns_to_extract]
    display(extracted_columns_df.head())
```

## Replace characters

### Subtask:
In each of the extracted columns ('name', 'a360_name', and 'referencia_name'), replace the '.' character with spaces.


**Reasoning**:
Iterate through the specified columns and replace the '.' character with a space in each of them, handling potential missing values.



```python
columns_to_clean = ['name', 'a360_name', 'referencia_name']

for file_path in excel_files:
    print(f"Processing file: {file_path}")
    df = pd.read_excel(file_path)
    extracted_columns_df = df[columns_to_extract]

    for col in columns_to_clean:
        extracted_columns_df[col] = extracted_columns_df[col].fillna('').astype(str).str.replace('.', ' ', regex=False)

    display(extracted_columns_df.head())
```

## Store or display results

### Subtask:
Store the processed data or display it.


**Reasoning**:
Display the full processed DataFrame as the task is to store or display the data, and displaying is the simpler option for this subtask.



```python
for file_path in excel_files:
    print(f"Processing file: {file_path}")
    df = pd.read_excel(file_path)

    columns_to_clean = ['name', 'a360_name', 'referencia_name']
    for col in columns_to_clean:
        if col in df.columns: # Check if column exists before cleaning
            df[col] = df[col].fillna('').astype(str).str.replace('.', ' ', regex=False)

    # Create the new filename
    file_dir = os.path.dirname(file_path)
    file_name_without_ext = os.path.splitext(os.path.basename(file_path))[0]
    new_file_path = os.path.join(file_dir, f"{file_name_without_ext}_processed_full.xlsx") # Changed filename to indicate full data

    # Save the processed DataFrame to a new Excel file
    df.to_excel(new_file_path, index=False)
    print(f"Processed full data saved to: {new_file_path}")
```
